```python
data_file_path = os.path.join(path,'confirmed-cases (1).csv') # เอาที่อยู่ของไฟล์เชื่อมกับที่อยู่ของไฟล์เรา โดยกำหนดชื่่อไฟล์ที่เราต้องการ
data_covid = pd.read_csv(data_file_path)
```

```python
data_covid
```

```python
datafn_1 = data_covid[['age','sex','risk']] # เลือก column ที่ต้องการใช้ ได้แก่ age,sex และ risk 
datafn_1
```

หาค่า missing ของ column ที่ต้องการใช้

```python
datafn_1[['age','sex','risk']].isnull().any() # หาค่า missing
```

จัดการแทนค่า missing

**Age**

```python
# แทนค่า missing ของ age ด้วยการเติมค่า mean(ค่ากลาง, ค่าเฉลี่ย)
datafn_1['age'].mean() # หาค่า mean ของ age
datafn_1age = datafn_1.fillna(value={'age':datafn_1['age'].mean()}) # แทนค่า mean ในค่า missing
datafn_1age
```

```python
 datafn_1age.isnull().any() # เช็คค่า missing ของ age
```

**sex**

```python
# แทนค่า missing ของ age ด้วยการเติมค่า mode 
datafn_1['sex'].mode() # หาค่า mode ของ age
```

จากการหา mode พบว่า เป็นเพศหญิงมากที่สุด ดังนั้นเราจะแทน missing 'sex' เป็นเพศหญิง

```python
datafn_1sex = datafn_1age.fillna(value={'sex':'หญิง'}) # แทน 'หญิง' ในค่า missing
datafn_1sex
```

```python
datafn_1sex.isnull().any() # เช็คค่า missing ของ sex
```

**risk**

```python
# แทนค่า missing ของ risk ด้วยการเติมค่า class ใหม่
datafn_1risk = datafn_1sex.fillna(value={'risk': 'ไม่ระบุ'}) # แทนค่า 'ไม่ระบุ' ในค่า missing
datafn_1risk
```

```python
datafn_1all = datafn_1risk
datafn_1all.isnull().any() # เช็คค่า missing
```

**เลือก class ของ column 'risk' ที่จะนำมาวาดกราฟ**

```python
datafn_1all['risk'].unique()
```

```python
datafn_class = datafn_1all[(datafn_1all['risk'] =='ร้านอาหาร')|(datafn_1all['risk'] =='ผับหัวหิน')|(datafn_1all['risk'] =='ตลาด')]
datafn_class
```

```python
datafn_1all[(datafn_1all['risk'] =='ตลาด')] # เช็คจำนวนข้อมูล column risk ที่มี ตลาด
```

```python
datafn_class['risk'].unique() # เช็คว่า class ที่เลือกมาถูกต้องหรือไม่
```

**แทนค่า risk เป็นตัวเลข เพื่อให้ง่ายต่อการวาดกราฟ**

```python
# แทนตัวหนังสือเป็นเลขด้วยคำสั่ง .replace() 
datafn_class = datafn_class.replace('ร้านอาหาร',1)
datafn_class = datafn_class.replace('ผับหัวหิน',2)
datafn_class = datafn_class.replace('ตลาด',3)
datafn_class
```

**แทนค่า sex เป็นตัวเลข เพื่อให้ง่ายต่อการวาดกราฟ**

```python
# แทนตัวหนังสือเป็นเลขด้วยคำสั่ง .replace() 
datafn_class = datafn_class.replace('ชาย',0)
datafn_class = datafn_class.replace('หญิง',50)
datafn_class
```

```python
pd.plotting.parallel_coordinates(datafn_class,'risk',colormap = 'Set1');
```
